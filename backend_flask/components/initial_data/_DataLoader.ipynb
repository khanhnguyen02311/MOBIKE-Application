```python
import pandas as pd
import numpy as np
```

```python
posts = pd.read_csv("Post.csv")
vehicleinfos = pd.read_csv("VehicleInfo.csv")
```

```python
posts.rename(columns={"ID_VehicleInfo": "infoid"}, inplace=True)
posts
```

```python
vehicleinfos
```

```python
completedposts = posts.set_index("infoid").join(vehicleinfos.set_index("infoid"))
```

```python
completedposts['Manufacture_year'] = completedposts['Manufacture_year'] // 100
completedposts['Cubic_power'] = completedposts['Cubic_power'].round()
```
