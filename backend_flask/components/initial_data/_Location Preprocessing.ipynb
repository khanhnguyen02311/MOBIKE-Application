```python
import pandas as pd
import numpy as np
from copy import deepcopy
```

```python
df = pd.read_excel('./Locations.xlsx', index_col=False)
```

```python
data = df.drop(labels=["Mã TP", "Mã QH", "Mã PX", "Cấp", "Tên Tiếng Anh"], axis=1).rename(columns={"Tỉnh Thành Phố":"City", "Quận Huyện":"District", "Phường Xã":"Ward"})
```

```python
data
```

```python
city_unique = data.loc[:,['City']].drop_duplicates()
district_unique = data.loc[:, ['District', 'City']].drop_duplicates()
ward_unique = data.loc[:, ['Ward', 'District', 'City']].drop_duplicates().dropna()
```

```python
city_unique = city_unique.reset_index().drop(columns=['index'])
district_unique = district_unique.reset_index().drop(columns=['index'])
ward_unique = ward_unique.reset_index().drop(columns=['index'])

city_unique['City_UID'] = city_unique.index + 1
district_unique['District_UID'] = district_unique.index + 1
ward_unique['Ward_UID'] = ward_unique.index + 1
```

```python
city_unique
```

```python
district_unique
```

```python
ward_unique
```

```python
city_table = deepcopy(city_unique)
district_table = district_unique.merge(city_unique, on='City', how='left').drop(labels=["City"], axis=1)
ward_table = ward_unique.merge(district_unique, on=['City','District'], how='left').drop(labels=['District', 'City'], axis=1)
```

```python
city_table.to_csv('Locations_City.csv', index=None)
district_table.to_csv('Locations_District.csv', index=None)
ward_table.to_csv('Locations_Ward.csv', index=None)
```
